```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data=pd.read_csv('Mobile_data.csv')
```

### Tasks
1. Remove the ₹ symbol from the 'Price' column
2. Remove the phrase 'Front Camera' and 'MP' in every attribute of the column 'Front Camera'
3. Remove all the phrases after 'mAh' in the battery column
4. Remove all the phrases after '(' in the Display column
5. Diaplay only Name(Colour) in the Name column
6. Remove 'RAM' term in the RAM column
7. Remove 'ROM' term in the ROM column

```python
data.head()
```

```python
data['Display']=data['Display'].apply(lambda x: x.split('cm')[0])
```

```python
data['Front Camera']=data['Front Camera'].apply(lambda x: x.split('MP')[0])
```

```python
data['Battery']=data['Battery'].apply(lambda x: x.split('mAh')[0])
```

```python
data['Price']=data['Price'].apply(lambda x: x.replace('?',''))
```

```python
data['RAM']=data['RAM'].apply(lambda x: x.split('RAM')[0])
```

```python
data['Internal storage']=data['Internal storage'].apply(lambda x: x.split('ROM')[0])
```

```python
data['Name']=data['Name'].apply(lambda x: x.split(',')[0])+')'
```

```python
result=data
result.to_csv("cleaned data.csv")
```
